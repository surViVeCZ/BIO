```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Create a list of image names
image_names = [f"{i}.jpg" for i in range(1, 9)]

# Define the plot with 2 rows
fig, axes = plt.subplots(2, len(image_names) // 2, figsize=(16, 6))

# Plot images in two rows
for i in range(len(image_names)):
    img = mpimg.imread(image_names[i])

    row = i // (len(image_names) // 2)  # Row index
    col = i % (len(image_names) // 2)  # Column index

    axes[row, col].imshow(img)
    axes[row, col].set_title(f"Iterace {i + 1}")
    axes[row, col].axis('off')

# Add labels underneath the second row
for j, ax in enumerate(axes[1]):
    ax.set_xlabel(f"Label {j + 1}")

plt.tight_layout()
plt.show()
```
